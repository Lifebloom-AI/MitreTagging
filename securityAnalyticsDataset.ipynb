```python
with open('savedsearches.conf', 'r') as file:
    content = file.read()
    # print(content[0:1000])
```

```python
ss = []
for segment in content.split('\n['):
  ss.append(segment)
```

```python
ss[2].split('\n')
```

```python
allContent = []
for search in ss[1:]:
  if 'action.notable =' in search or 'action.correlationsearch' in search:
    content = {}
    try:
      FP = search.split('known_false_positives = ')[1].split('\n')[0]
    except:
      FP = ''
    try:
      story = search.split('.analytic_story = ')[1].split('\n')[0]
    except:
      story = ''
    try:
      dataSource = search.split('.data_models = ')[1].split('\n')[0]
    except:
      dataSource = ''
    try:
      eli5 = search.split('eli5 = ')[1].split('\n')[0]
    except:
      eli5 = ''
    content[search.split(']\n')[0]] = {'description':search.split('\ndescription = ')[1].split('\n')[0],
                                      'search':search.split('\nsearch = ')[1].split('\n')[0],
                                       'knownFP':  FP, 'story': story, 'eli5': eli5,
                                       'dataSource': dataSource
                                       }
    allContent.append(content)
```
